## Задание 3

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json`.

Выполнить сохранение в формат CSV. Прочитать полученный файл. Сравнить с исходным.


````python
import pandas as pd
import json
import urllib

url = "https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json"
data = urllib.request.urlopen(url)

json_dict = json.load(data)

df_csv = pd.json_normalize(json_dict, record_path='bio', meta=['name', 'years', 'genre', 'nationality'])

df_csv

df_json = pd.read_json(url)

output_file = 'artists.csv'

df_json.to_csv(output_file, index=False)

if df_json.equals(df_csv):

    print("Сравнение исходных и загруженных данных: Cовпадают")

else:
    print("Сравнение исходных и загруженных данных: Не совпадают")

df_csv
````
