```python

# Подключаем библиотеки
import pandas as pd # Работа с набором данных
import numpy as np # Линейная алгебра
# Визуализация данных
import seaborn as sns
import matplotlib.pyplot as plt

# отключение информационных сообщений
from warnings import filterwarnings
filterwarnings(action='ignore')

```

```python
# читаем данные
url = 'https://raw.githubusercontent.com/yakushinav/omo/refs/heads/main/data/SomervilleHappinessSurvey2015.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.plot(kind='scatter',x='X2', y='X5')
```

```python
sns.FacetGrid(data, hue = 'D') \
    .map(plt.scatter, 'X2','X4') \
    .add_legend()
```

```python
n_bins = 20
fig, axs = plt.subplots(3, 2)
axs[0,0].hist(data['X1'], bins = n_bins);
axs[0,0].set_title('X1');
axs[0,1].hist(data['X2'], bins = n_bins);
axs[0,1].set_title('X2');
axs[1,0].hist(data['X3'], bins = n_bins);
axs[1,0].set_title('X3');
axs[1,1].hist(data['X4'], bins = n_bins);
axs[1,1].set_title('X4');
axs[2,0].hist(data['X5'], bins = n_bins);
axs[2,0].set_title('X5');

fig.tight_layout(pad=1.0);
```

```python
data.info()
```

```python
sns.boxplot(x='D', y='X4', hue='D', data=data)
```

```python
sns.FacetGrid(data, hue="D") \
   .map(sns.kdeplot, "X5") \
   .add_legend()
```

```python

sns.pairplot(data, hue='D', size=3)
```

```python
corrmat = data.corr(numeric_only=True)
sns.heatmap(corrmat, annot = True, square = True);
```
